```python
import os
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
from nltk.tokenize import RegexpTokenizer
import math
import numpy as np
import pandas as pd
```

# IMPLEMENTATION FUNCTIONS(1,2,3,4)
Contains                                                                         
1.)Cleaning of data                                                               
2.)Creation of Dictionaries

```python
#1
#ACCESSING THE FILE AND TOKENIZING WORDS OUT OF IT
#OPERATION:-TOKENIZE,READ_FILE

def tokenize_word(file):
    t=word_tokenize(file)
    tokenizer = RegexpTokenizer(r'\w+')
    tokens=tokenizer.tokenize(file)
    print("tokenize_word=",len(tokens))
 
    return tokens
```

```python
#2
#CONVERTING TEXT INTO LOWER_CASE

def toLowerCase(file):
    
    tokens=tokenize_word(file)
    for i in range(len(tokens)):
        tokens[i]=tokens[i].lower()
    #print("to lower case=",len(tokens))
    
    return tokens
```

```python
#3
#REMOVING STOP WORDS
#AND
#NUMERIC STRINGS
#AND
#STRINGS OF LENGTH LESS THAN 2 ARE REMOVED

def cleanData(file):
    tokens=toLowerCase(file)
    arr=[]
    stop_words=set(stopwords.words("english"))
    for word in tokens:
        if word not in stop_words and not word.isnumeric() and len(word)>2:#
            arr.append(word)
    #print("cleanData=",len(arr))  
    
    return arr
```

```python
#4
#CREATE DICTIONARY FOR EVERY DOCUMENT CONTAINING UNIQUE WORDS PER DOCUMENT:-d={}
#CREATE DICTIONARY CONTAINING ALL DICTINCT WORDS FROM ALL DOCUMENTS:-d_all={}

def createDict(file,d,filename,sub):
    arr=cleanData(file) #CALLING CleanData Method #Getting array of all words in file #Repetition may be found
    
    d[filename]={}
    d[filename]["Class"]=sub
    d[filename]["Words"]={}
    
    #print("set(arr)=",len(set(arr)))
    for word in set(arr):
            d[filename]["Words"][word]=0
            if word not in d_all:
                d_all[word]=0
            l.append(word)

    for i in set(arr):
        for j in arr:
            if i==j:
                d[filename]["Words"][i]+=1
                d_all[i]+=1      
    

```

```python
#5
#MAIN METHOD

#READING FILE
#TOKENIZING FILE
#CONVERTING FILE TO LOWER CASE
#REMOVING STOP WORDS,NUMERIC STRINGS AND STRINGS LESSER THAN LENGTH 2

d_class={}
d={}                  #CONTAINS ALL DOCUMENTS NAME WITH ALL UNIQUE WORDS IN THAT DOCUMENT
d_all={}              #CONTAINS ALL DICTINCT WORDS OUT OF EACH DOCUMENT WITH COUNT #NO REPETITION OF WORDS #ALL UNIQUE WORDS
l=[]                  #contains dictinct words of each document #summed together #hence might have words repeating #NON USEFUL
csub=0
cfile=0
os.chdir("C:\\Users\\hp\\Coding Ninjas\\Lectures\\Lect 9(Text Classification)\\Dataset\\20_newsgroups_train")
for subfolders in os.listdir():                 #LOOP 1 #LOOPS ALL SUBFOLDERS
    path=("C:\\Users\\hp\\Coding Ninjas\\Lectures\\Lect 9(Text Classification)\\Dataset\\20_newsgroups_train")
    csub+=1
    print(csub,".)","subfolders=",subfolders)
    os.chdir(os.path.join(path,subfolders))
    for filename in os.listdir():               #LOOP 1.1 #LOOPS ALL FILES IN SUBFOLDERS
        if(filename!=".DS_Store"):
            cfile+=1
            print(cfile,".)","file=",filename)
            file=open(filename,"r+")            #OPENS FILE
            file=file.read()                    #CONVERTS FILE INTO STRING


            createDict(file,d,filename,subfolders) #MAIN METHOD CALLING METHODS:-1,2,3,4

         
```

```python
#6
#CREATING NEW DICTIONARY:-dt={}

#STORES JUST THE TOP FEATURES FOUND OUT OF ALL DOCUMENTS
#TOP FEATURES:-WORDS WITH MAX FREQUENCY IN ALL DOCUMENTS

#ROW=DOCUMENTS
#COLUMNS=TOP FEATURES(WORDS WITH MAX FREQ.)
#VALUES=NO OF TOP FEATURE PRESENT IN THAT DOCUMENT i.e.VALUES=dt[DOCUMENT][TOP FEATURE]

all_unique_features=sorted(d_all, key=d_all.get, reverse=True)
top_features=all_unique_features[0:6500]

dt={}
columnCount=0
l=[]
l2=[]
for doc in d.keys():
    columnCount+=1
    dt[doc]={}
    dt[doc]["Words"]={}
    dt[doc]["Class"]={}
    dt[doc]["Class"]=d[doc]["Class"]
    l2.append(d[doc]["Class"])
    for word in top_features:
        if word in d[doc]["Words"]:
            dt[doc]["Words"][word]=d[doc]["Words"][word]
            l.append(d[doc]["Words"][word])
            
        else:
            dt[doc]["Words"][word]=0
            l.append(0)


#len(set(d_all))
#all_unique_features
d_all
```

```python
#7
#MAKING DICTIONARY d_class
#STORES THE COUNT OF N TOP FEATURES IN A PARTICULAR CLASS
#d_class STORES ALL INFORMATION AND TOP N FEATURES
#ABOUT ALL CLASSES


l1=[]
for i in set(dt.keys()):                    #LOOP 1   #LOOPS IN DOCUMENTS #AND FINDS ALL POSSIBLE CLASSES IN ALL DOCUMENTS
    l1.append(dt[i]["Class"])

l1=set(l1)
d_class={}
for i in l1:                                #LOOP 2   #LOOPS THROUGH ALL CLASSES 
    d_class[i]={}
    d_class[i]["Total Words"]=0
    d_class[i]["Total Files"]=0
    d_class[i]["Words"]={}
    for j in set(dt["38839"]["Words"]):     #LOOP 2.1 #LOOPS THROUGH TOP N FEATURES(WORDS) #SETS EACH VALUE 0
        d_class[i]["Words"][j]=0

for i in set(dt.keys()):                    #LOOP 3 #LOOPS ALL DOCUMENTS
    class_name=dt[i]["Class"]
    d_class[class_name]["Total Files"]+=1
    for j in set(dt[i]["Words"]):           #LOOP 3.1 #LOOPS TOP N FEATURES 
        d_class[class_name]["Words"][j]+=dt[i]["Words"][j]
        d_class[class_name]["Total Words"]+=dt[i]["Words"][j]

        
d_class
```

# TESTING                                                                          
Calculates SCORE

```python
#1
#TESTING THE DATA AND CALCULATING THE SCORE

d1={}
csub=0
cfile=0
os.chdir("C:\\Users\\hp\\Coding Ninjas\\Lectures\\Lect 9(Text Classification)\\Dataset\\20_newsgroups_test")
positive_result,negative_result=0,0
for sub in os.listdir():                       #LOOP 1 #FOR SUBFOLDERS
    path1="C:\\Users\\hp\\Coding Ninjas\\Lectures\\Lect 9(Text Classification)\\Dataset\\20_newsgroups_test"
    path=os.path.join(path1,sub)
    os.chdir(path)
    
    for filename in os.listdir():              #LOOP 1.1 #FOR EACH DOCUMENT IN EVERY SUBFOLDERS
        cfile+=1
        print(cfile,".)","file=",filename)
        d1[filename]={}
        d1[filename]["output"]=sub
        
        if(filename!=".DS_Store"):
            file=open(filename,"r+")           #OPENS EACH FILE RETUNS POINTER
            file=file.read()                   #CONVERTS EACH FILE TO STRING

            p={}                               #STORES THE PROBABILITY OF EACH CLASS #FOR EACH DOCUMENT #INDIVIDUALLY
            max1=0                             #STORES THE MAX PROBABILITY
            max_class=0                        #STORES NAME OF MAX CLASS
            Total_files_all=0                  #STORES TOTAL NUMBER OF FILES IN TESTING
            arr=cleanData(file)                #METHOD 4(Clean Data) IS CALLED TO CLEAN TEST DOCUMENT "docx"
            arr=set(arr)                       #RETURNS TOKENIZED CLEANED WORDS
            
            for cl in d_class.keys():          #LOOP 1.2   #LOOPS ALL CLASSES
                p[cl]=1
                for word in arr:               #LOOP 1.2.1 #LOOPS ALL WORDS IN A PARTICULAR CLASS
                    Total_files_all+=d_class[cl]["Total Files"]
                    if word in d_class[cl]["Words"]:
                        total_words=d_class[cl]["Total Words"]#total_words HOLDS THE COUNT OF ALL WORDS IN THAT PARTICULAR CLASS
                        p[cl]+=math.log((d_class[cl]["Words"][word]+1)/total_words)#CALCULATING PROBABILITY FOR EVERY WORD

            for cl in set(p):                  #LOOP 1.3   #LOOPS PROBABILITIES OF ALL CLASSES 
                p[cl]+=math.log(d_class[cl]["Total Files"]/Total_files_all)
                max1=p[cl]                     
                max_class=cl

            #-------------------------------------------------------------------------------------------------------------------
            #PREDICTION PART
            #DECIDES CLASS WITH MAX PROBABILITY
            #AND PREDICTS IT AS THE RESULT
            
                       
            for i in set(p):                    #LOOP 1.4 #LOOP FINDS CLASS WITH MAX PROBABILITY
                if p[i]>max1:
                    max1=p[i]
                    max_class=i

            d1[filename]["predicted"]=max_class
            print("original=",sub,"predicted=",max_class)
            
            
            #-------------------------------------------------------------------------------------------------------------------
            #NOTING WHICH RESULTS COME OUT TO BE POSITIVE AND WHICH ALL COME OUT TO BE NEGATIVE
            #NEEDED TO CALCULATE SCORE
            #ALSO MAINTAINS COUNT OF POSITIVE AND NEGATIVE RESULTS
            #TO CALCULATE RESULTS
            
            
            if d1[filename]["predicted"]==d1[filename]["output"]:
                positive_result+=1
            else:
                negative_result+=1
                
    print()
```

# Self-Implementation Score

```python
#SCORE

score=(positive_result/(positive_result+negative_result))
selfscore=score*100
selfscore
```

```python
#COMPARISON FUNCTION
#BETWEEN INBUILT AND SELF-IMPLEMENTED SCORE

def comparison(selfscore,skscore):
    s=abs(selfscore-skscore) 
    if selfscore>skscore:
        print("Self Implemented Code Has Better Accuracy By : ", end=" ")
        print(s, end="") 
        print("%")
    elif selfscore<skscore:
        print("SkLearn Implemented Function Has Better Accuracy By : ", end=" ")
        print(s, end="")
        print("%")
    else:
        print("Both Self Implemented Code And SkLearn Implemented Function Have Equal Accuracy")
```
